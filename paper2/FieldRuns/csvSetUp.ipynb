# Set up CSV with run data for Field Runs

```python
import numpy as np 
import csv
import itertools
```

## QD runs

- 5 losses
- 5 fields
- 3 NDBPs
- 3 ngs
- 2 seeds

```python
# Generate all combinations
all_rows = np.array(list(itertools.product(range(5), range(5), range(3), range(3), range(2))))

header = ['LOSS_INDEX', 'FIELD_INDEX', 'NDBP_INDEX', 'NG_INDEX', 'SEED']

with open('QD_arrayInputs.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    writer.writerows(all_rows)
```
